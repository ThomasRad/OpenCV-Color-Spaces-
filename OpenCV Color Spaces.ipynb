### Color Spaces

```python
# Import OpenCv

import cv2 as cv

# Read the image 

img = cv.imread('/Users/tradk/dogs-vs-cats/train/cat.1.jpg')
```

```python
# Convert to Grayscale 

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Display the image to make sure. 

cv.imshow('Gray Image',gray)
cv.waitKey(0)
```

```python
# Converting to HSV

hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)

#Display the image. 

cv.imshow("HSV Image", hsv)
cv.waitKey(0)
```

```python
# LAB Color Space. 

lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)

#Display the image. 

cv.imshow('LAB COLOR', lab)
cv.waitKey(0)
```

```python
# Import package 

import matplotlib.pyplot as plt
```

```python
# Displaying images in packages outside of OpenCV

plt.imshow(img)
plt.show()
```

```python
# Converting BGR to RGB

rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Display in OpenCV

cv.imshow('RGB', rgb)

# Display in Matplot

plt.imshow(rgb)
plt.show()

cv.waitKey(0)
```
